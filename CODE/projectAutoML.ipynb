```
# restart of runtime is required to run it first time

import autosklearn.classification
import sklearn.model_selection
import sklearn.datasets
import sklearn.metrics
import pickle
```

```
import matplotlib
import numpy as np
import pickle

with open('/content/gdrive/MyDrive/Colab Notebooks/DATA_RATING/Xtr.pickle', 'rb') as handle:
    Xtr=pickle.load( handle)
with open('/content/gdrive/MyDrive/Colab Notebooks/DATA_RATING/Xts.pickle', 'rb') as handle:
    Xts=pickle.load(handle)
with open('/content/gdrive/MyDrive/Colab Notebooks/DATA_RATING/ytr.pickle', 'rb') as handle:
    ytr= pickle.load(handle)
with open('/content/gdrive/MyDrive/Colab Notebooks/DATA_RATING/yts.pickle', 'rb') as handle:
    yts= pickle.load( handle)
```

```
Xtr.shape
```

```
automlNoXc = autosklearn.classification.AutoSklearnClassifier(time_left_for_this_task=60*60) #Auto-sklearn searches pipelines for 60 minutes

automlNoXc.fit(Xtr, ytr)

yhat = automlNoXc.predict(Xts.squeeze())

print("Accuracy score", sklearn.metrics.accuracy_score(yts, yhat))

with open('/content/gdrive/MyDrive/Colab Notebooks/automlProject.pickle', 'wb') as handle:
    pickle.dump(automlNoXc, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```
with open('/content/gdrive/MyDrive/Colab Notebooks/automlProject.pickle', 'rb') as handle:
    automlProject=pickle.load( handle)
automlProject.show_models()
```
